```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

EDA IRIS

```python
print(sb.get_dataset_names())
df = sb.load_dataset("iris")
# display(df)
print(df.shape, df.columns, df["species"].value_counts())

# Scatter plot basic
df.plot(kind = "scatter", x= "sepal_length", y="sepal_width")
plt.show()

# Scatter plot advanced using seaborn
sb.set_style("whitegrid")
sb.FacetGrid(df, hue = "species", height = 4)\
.map(plt.scatter, "sepal_length", "sepal_width")\
.add_legend()

```
